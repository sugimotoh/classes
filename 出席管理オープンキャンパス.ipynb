```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# 出席データを格納するリスト
event_data = []

# データを記録する関数
def record_school_data(button):
    school_name = school_input.value
    num_students = students_input.value
    gender = gender_input.value
    remarks = remarks_input.value

    if not school_name or num_students is None or not gender or not remarks:
        with output:
            clear_output()
            print("すべてのフィールドを入力してくださいにゃー")
        return

    event_data.append({'School': school_name, 'Number_of_Students': num_students, 'Gender': gender, 'Remarks': remarks})

    with output:
        clear_output()
        print("データが記録されましたにゃー")
        print(pd.DataFrame(event_data))

    # 入力フィールドをクリア
    school_input.value = ''
    students_input.value = 0
    gender_input.value = '男'
    remarks_input.value = ''

# インタラクティブなウィジェットを作成
school_input = widgets.Text(description="高校名", placeholder="高校名を入力してください")
students_input = widgets.IntText(description="人数", value=0, placeholder="人数を入力してください")
gender_input = widgets.Dropdown(options=['男', '女', '男女混合', 'ご父母同伴','その他'], description="性別")
remarks_input = widgets.Text(description="備考", placeholder="備考を入力してください")
submit_button = widgets.Button(description="送信")
output = widgets.Output()

# ボタンがクリックされたときに関数を呼び出す
submit_button.on_click(record_school_data)

# ウィジェットを表示
display(school_input, students_input, gender_input, remarks_input, submit_button, output)

# 最後にCSVに保存するための関数を作成
def save_to_csv():
    df = pd.DataFrame(event_data)
    df.to_csv('/content/event_attendance.csv', index=False)
    with output:
        clear_output()
        print("データが保存されましたにゃー")
        print(df)

# 保存ボタンを作成して表示
save_button = widgets.Button(description="保存")
save_button.on_click(lambda x: save_to_csv())
display(save_button)
```
