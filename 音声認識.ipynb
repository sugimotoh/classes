```python
import gradio as gr
import whisper
import torch

# Whisperモデルのロード
model = whisper.load_model("base")

def speechRecognitionModel(audio):
    if audio is None:
        return "No audio provided"

    # 音声ファイルのロード
    audio = whisper.load_audio(audio)
    audio = whisper.pad_or_trim(audio)

    # メルスペクトログラムの生成
    mel = whisper.log_mel_spectrogram(audio).to(model.device)

    # デコードオプションの設定
    options = whisper.DecodingOptions(language="ja")

    # 音声をテキストに変換
    result = whisper.decode(model, mel, options)

    return result.text

# Gradioインターフェースの設定
interface = gr.Interface(
    fn=speechRecognitionModel,
    inputs=gr.Audio(),
    outputs=gr.Textbox(),
    live=True,
    title="Whisper Speech Recognition Model",
    description="日本語の音声を認識するためのモデル"
)

# インターフェースの起動
interface.launch()
```
