```
import pandas as pd
import requests
from io import BytesIO
import numpy as np
```

```
# エクセルファイルのURLを指定する。
url = "https://www.nstac.go.jp/sys/files/SSDSE-D-2023.xlsx" #SSDSE-基本素材（SSDSE-E） 都道府県
url02 = "https://www.nstac.go.jp/sys/files/SSDSE-F-2023.xlsx"  #SSDSE-気候値（SSDSE-F）
url03 = "https://www.nstac.go.jp/sys/files/SSDSE-C-2024.xlsx" #SSDSE-家計消費（SSDSE-C）
#url = "https://www.nstac.go.jp/sys/files/SSDSE-A-2023.xlsx" #市区町村

# 指定したファイルをダウンロードする。
response01 = requests.get(url)
response02 = requests.get(url02)
response03 = requests.get(url03)

# エクセルファイルを読み込む
#　BytesIOにより、ダウンロードしたバイトデータを直接メモリ上で処理する
#　header=1の指定により、データフレームの2レコード目（2行目）が変数名となる。
data01 = pd.read_excel(BytesIO(response01.content), header=1)
# データフレームに変換する
data01=pd.DataFrame(data01)
# データの内容を表示
data01.head()

data02 = pd.read_excel(BytesIO(response02.content), header=1)
data02=pd.DataFrame(data02)
data03 = pd.read_excel(BytesIO(response03.content), header=1)
data03=pd.DataFrame(data03)
```

```
import pandas as pd
import numpy as np

# データフレームを作成または読み込む
# ここでは例としてランダムなデータフレームを作成する
np.random.seed(0)
#df010203 = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])

# 相関行列を計算
correlation_matrix = df_dropped.corr()

# 相関が高い組み合わせを抽出するためのしきい値を設定
threshold = 0.65
# 相関行列の上三角行列を取得
upper_triangle = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))

# 相関がしきい値を超える組み合わせを見つける
high_corr_vars = [(correlation_matrix.index[i], correlation_matrix.columns[j], upper_triangle.iat[i, j])
                  for i, j in zip(*np.where((upper_triangle >= threshold) | (upper_triangle <= -threshold)))]

# 見つけた高相関の組み合わせを表示
for var1, var2, corr in high_corr_vars:
    print(f'Pair: ({var1}, {var2}) - Correlation: {corr}')
```

```
from datetime import datetime

# 測定された時刻
measured_time_str = data_filtered03['仕事からの帰宅時間']
measured_time = datetime.strptime(measured_time_str, '%H:%M:%S')

# 比較する時間（19:00:00）
compare_time_str = '19:00:00'
compare_time = datetime.strptime(compare_time_str, '%H:%M:%S')

# 時間の差分を計算
time_difference = compare_time - measured_time

# 差分を分に変換
time_difference_minutes = time_difference.total_seconds() / 60

# 結果を表示
print(f"測定された時刻: {measured_time_str}")
print(f"19:00:00からの時間差: {time_difference_minutes} 分")
```
