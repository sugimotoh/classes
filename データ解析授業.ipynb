```
import pandas as pd
import requests
from io import BytesIO
import numpy as np
```

```
# エクセルファイルのURLを指定する。
#url = "https://www.nstac.go.jp/sys/files/SSDSE-D-2023.xlsx" #都道府県
url = "https://www.nstac.go.jp/sys/files/SSDSE-D-2023.xlsx" #SSDSE-基本素材（SSDSE-E） 都道府県
#url = "https://www.nstac.go.jp/sys/files/SSDSE-A-2023.xlsx" #市区町村

# 指定したファイルをダウンロードする。
response = requests.get(url)

# エクセルファイルを読み込む
#　BytesIOにより、ダウンロードしたバイトデータを直接メモリ上で処理する
#　header=1の指定により、データフレームの2レコード目（2行目）が変数名となる。
data = pd.read_excel(BytesIO(response.content), header=1)
# データの内容を表示
data=pd.DataFrame(data)
# データの内容を表示
data.head()
```
