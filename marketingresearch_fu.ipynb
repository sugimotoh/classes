#  11/30 9:00-9:30**クロス集計分析**

```python
import pandas as pd
crossed00=pd.crosstab(DF11['アンケート_クチコミタイプ'], DF11['Q6_クチコミ評価'])
crossed00
```

# **chi-square test（統計的検定）**

```python
import scipy
from scipy import stats
x2, p, dof, expected = stats.chi2_contingency(crossed00)
print("p-value %(p)s" %locals() )

if 0 < p < 0.01: print("1%水準有意")
elif 0.01 <= p <= 0.05: print("5%水準有意")
elif 0.05 <= p <= 0.10: print("10%水準有意")
else: print("not significant")
```

```python
crossed01=pd.crosstab(DF11['アンケート_クチコミタイプ'], DF11['Q7_実写映画_視聴意向'])
crossed01
```

```python
x2, p, dof, expected = stats.chi2_contingency(crossed01)
print("p-value %(p)s" %locals() )

if 0 < p < 0.01: print("1%水準有意")
elif 0.01 <= p <= 0.05: print("5%水準有意")
elif 0.05 <= p <= 0.10: print("10%水準有意")
else: print("not significant")
```
