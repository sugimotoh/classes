```
import pandas as pd
import requests
from io import BytesIO
import numpy as np
```

```
# エクセルファイルのURLを指定する。
#url = "https://www.nstac.go.jp/sys/files/SSDSE-D-2023.xlsx" #都道府県
url = "https://www.nstac.go.jp/sys/files/SSDSE-D-2023.xlsx" #SSDSE-基本素材（SSDSE-E） 都道府県
#url = "https://www.nstac.go.jp/sys/files/SSDSE-A-2023.xlsx" #市区町村

# 指定したファイルをダウンロードする。
response = requests.get(url)

# エクセルファイルを読み込む
#　BytesIOにより、ダウンロードしたバイトデータを直接メモリ上で処理する
#　header=1の指定により、データフレームの2レコード目（2行目）が変数名となる。
data = pd.read_excel(BytesIO(response.content), header=1)
# データフレームに変換する
data=pd.DataFrame(data)
# データの内容を表示
data.head()
```

```
# "全国"のレコードを除外
data_filtered01 = data[data["都道府県"] != '全国']
data_filtered02 = data_filtered01[data_filtered01["男女の別"] == '0_総数']
data_filtered02.head()
```

```
import plotly.express as px

# Plotlyを使って散布図を作成する
fig = px.scatter(data_filtered02, x='商業実務・ビジネス関係', y='サイクリング', color='都道府県')

# 背景を白くし、幅と高さを設定
fig.update_layout(
    plot_bgcolor='white',  # プロット領域の背景色
    paper_bgcolor='white',  # 全体の背景色
    width=700,  # 図の横幅（ピクセル）
    height=600  # 図の縦幅（ピクセル）
)
# 散布図を表示
fig.show()
```

```
# "0_総数"のレコードを除外
data_filtered03 = data_filtered01[data_filtered01["男女の別"] != '0_総数']
data_filtered03.head()
```
